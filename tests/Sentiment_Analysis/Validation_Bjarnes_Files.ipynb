```python
import csv
import numpy as np
import pandas as pd
import re
import emoji
```

```python
list_error = []
path_ori = "/Users/tobias/Dev/FOM/Master_Thesis/data/Final_DE/Validation/"
path_en = "/Users/tobias/Dev/FOM/Master_Thesis/data/Final_EN/Validation/"
path_bj = "/Users/tobias/Dev/FOM/Master_Thesis/data/Final_DE/Validation_Bjarne/"

def check_files(lang, path = path_ori):
    for file_nbr in range(0,15):
        print("File_Nbr.: ", file_nbr)
        list_error_files = []
        
        file_ori_path = path + str(lang) + "_validation_" + str(file_nbr) + ".csv"
        file_bj_path = path_bj + str(lang) + "_validation_" + str(file_nbr) + ".csv"
        
        print("Path Validation: ", file_ori_path)
        print("Path BJ: ", file_bj_path)
        
        df_ori = pd.read_csv(file_ori_path, sep=";")
        df_bj = pd.read_csv(file_bj_path, sep=";")
        
        print("Length DF_ORI: ", len(df_ori))
        print("Length DF_BJ: ", len(df_bj))
        
        counter = 0
        for each in range(len(df_ori)):
            try:
                text_ori = re.sub('[^A-Za-z0-9]+', '', df_ori.iloc[each]["text"])
                text_bj = re.sub('[^A-Za-z0-9]+', '', df_bj.iloc[each]["text"])
                
                if text_ori == text_bj:
                    continue
                else:
                    list_error_files.append(each)
                    counter =counter + 1
            except Exception:
                print(each, "ERROR")
        list_error.append(list_error_files)
        print(counter, "- Errors in the file: ", file_nbr)
        print("____________________________\n")

check_files("de")
check_files("en", path_en)
```
